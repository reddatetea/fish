Raw.xlsx

1.添加1列: Num2, 仅选出Num=1时，对应的Name, 其余为空值

2.添加1列: Item2, 按照Name的分类，将Item连接起来，排除空值

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_clipboard()
df
```

```python
Item2 = df.groupby('Name').Item.transform(lambda x: ','.join(i for i in x if i is not np.nan))
Item2
```

```python
(
    df.assign(Num2=np.where(df.Num == 1, df.Name, ''))
    .assign(Item2=Item2)
)
```
