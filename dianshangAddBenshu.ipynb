```python
import re
import pandas as pd
import numpy as  np
```

```python
fname = r"F:\a00nutstore\008\zw08\电商\电商7-16订单表正.xlsx"
sheet_name = r'Sheet1'
skiprows = 0
```

```python
df = pd.read_excel(fname,sheet_name = sheet_name,skiprows = 0)
df
```

```python
# df.dropna(subset = ['订单号'],inplace = True)
df['规格名称1'] = df['规格名称']
```

```python
df['规格名称1'] = df['规格名称1'].str.replace('(10本)','10本').str.replace('五本','5本').str.replace('三本','3本').str.replace('(5本)','5本').str.replace('10本','(10本)').str.replace('20本','(10本)').str.replace('4本','(4本)').str.replace('5本','(5本)').str.replace('3本','(3本)')
```

```python
def addBen(string):
    # string = r'（15本)'
    pattern = r'[(（](?P<ben>\d+)本[)）]'
    regexp = re.compile(pattern)
    mat = regexp.search(string)
    if mat:
        ben = int(mat.group('ben'))
    else :
        ben = 1
    return ben
```

```python
df['ben'] = df['规格名称1'].map(addBen)
df
```

```python
df.to_excel('220.xlsx',index = False)
```
