```python
'''
tcloud中用销售订单生成税局发票模板
'''
import os
import re
import easygui
import openpyxl
import numpy as np
import pandas as pd
```

```python
def danjuChuli(df):
    df = df.dropna(how='all', axis=0)
    cols = [i for i in df.columns.to_list() if 'Unnamed' not in i]
    df = df[cols]
    # get maxrows
    for label, ser in df.items():
        for num, x in enumerate(ser):
            if isinstance(x, str):
                if '合计' in x:
                    index = num
                    print(index)
                    break
    df = df.iloc[:index]
    df = df.dropna(how='all', axis=1)
    return df
def chuliMingchen(d):
    qian = d['存货名称'].str.split('型').str[0]
    hou = d['存货名称'].str.split('型').str[1]
    return qian,hou
```

```python
#读取发票模板
fname_fapiao = r"F:\repos\fish\发票模板.xlsx"
df_fapiao = pd.read_excel(fname_fapiao,header = 2,dtype = {'商品和服务税收分类编码':'str'})
df_fapiao
```

```python
fname_dingdan = r"F:\repos\fish\肥猫销售订单SO-2024-04-0082.xlsx"
df_dingdan = pd.read_excel(fname_dingdan,sheet_name = 0,header = 8)
df_dingdan
```

```python
df_dingdan1 = danjuChuli(df_dingdan)
qian,hou = chuliMingchen(df_dingdan1)
df_dingdan1['项目名称'] = df_dingdan1['货号'] + '-' + hou
df_dingdan1['规格型号'] = qian
df_dingdan1['商品和服务税收分类编码'] = '1060202010000000000'	
df_dingdan1['商品数量'] = df_dingdan1['数量（本）']
df_dingdan1['商品单价'] = df_dingdan1['单价']
df_dingdan1['金额'] = df_dingdan1['金额（元）']
df_dingdan1['单位'] = '本'
df_dingdan1['税率'] = 0.01
df_dingdan1['折扣金额'] = ''
df_dingdan1['优惠政策类型'] = ''

df_dingdan2 = df_dingdan1[['项目名称',
 '商品和服务税收分类编码',
 '规格型号',
 '单位',
 '商品数量',
 '商品单价',
 '金额',
 '税率',
 '折扣金额',
 '优惠政策类型']]
df_dingdan2
```

```python
fname_result = r"F:\repos\fish\发票模板 - mm.xlsx"
# df_dingdan2.to_excel(fname_result,sheet_name = 0,startrow = 2,header = None,index = False)
wb = openpyxl.load_workbook(fname_result)
ws0 = wb.worksheets[0]
with pd.ExcelWriter(fname_result, engine='openpyxl')  as writer:
    writer.book = wb
    writer.sheets = dict((ws.title, ws) for ws in wb.worksheets)
    df_dingdan2.to_excel(writer, ws0, header=None, index=False, startrow=2)
writer.save()
```
